```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import statistics as stat

#busco el nombre de la base de datos que voy a utlizar
%ls datasets
```

```python
#se llama a la base de datos que se quiere utlizar y se observa los primeros renglos de la base de datos
df=pd.read_csv("datasets/antropometria-dataset.csv")
df.head(4)
```

```python
#¿Cúal es el promedio y moda de peso?
#Es importante saber el promedio de peso que tiene la población,lo cual no ayudará a saber si estan dentro del sobrepeso o no.
a=df.loc[:,"peso"]
np.sum(a)/len(a) #promedio de pesos
```

```python
stat.mode(a) #moda
#esto nos dice que la mayoria de la personas pesan 63.8
```

```python
#¿Qué género fue el más entrevistado?
sns.countplot(x="sexo",
             data=df)
plt.xticks(rotation=0)
plt.yticks(rotation=360)
plt.show()
```

```python
#¿El peso afecta la talla de las prendas dependiendo del género?
sns.lmplot(x="peso",
          y="talla",
           hue="sexo",
          markers=["x","o"],
          fit_reg=False,
          data=df)
plt.title("Diagrama de dispersion de talla con respecto al peso por cada sexo")
plt.show()
```

```python
#¿Será que los dos géneros teniendo el mismo peso poseen la misma magnitud de caderas?
import seaborn as sns
sns.lmplot(x="peso",
          y="cadera",
          data=df,
          fit_reg=True,
          hue="sexo")
plt.title("Scaterr plot de cadera con respecto al peso")
plt.show()
```

```python
#¿Acaso adelgazan las personas en el transcurso de sus vidas?
x=df["edad"].values
y=df["peso"].values

plt.scatter(x,y,c="y")
plt.title("Diagrama de dispersion de peso con respecto a la edad")

plt.xlabel("Edad")
plt.ylabel("Peso")

plt.show()
```

```python
#¿Si engordamos, la cintura crece mucho?

x=df["peso"].values
y=df["cintura"].values

plt.scatter(x,y,c="r")
plt.title("Diagrama de dispersion de cintura con respecto al peso")

plt.xlabel("Peso")
plt.ylabel("Cintura")

plt.show()
```
