```python
import numpy as np
import pandas as pd
df=pd.read_csv("datasets/Nuevo_CANADA.csv")
df
```

```python
df.drop([195,196],axis=0,inplace=True)
```

```python
df.drop(["Type","Coverage","AREA","REG","DEV"],
       axis=1,
       inplace=True)
df.rename(columns={"OdName":"country",
                  "AreaName":"Continent",
                  "Regname": "region"},
         inplace=True)
```

```python
df.column=list(map(str,df.columns))
df.set_index("country",inplace=True)
years=list(map(str,range(1980,2014)))

new_df= df[years]
new_df
```

```python
df.head()
```

```python
new_df["Total"]=new_df.iloc[:,:].sum(axis=1)
new_df
```

```python
new_df=df.groupby("Continent",
                 axis=0).sum()
new_df
```

```python
new_df=new_df[years]
new_df["Total"]=new_df.iloc[:,:].sum(axis=1)
new_df
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
color_list=["gold","yellowgreen", "lightcoral", "lightskyblue",
            "lightgreen","pink"] #los colores que quieres
exp_list=[0.1,0.2,0,0,0.2,0.3] #para la separaciÃ³n entre cosas

new_df["Total"].plot(kind="pie",
                    figsize=(15,6),
                    autopct="%10.1f%%",
                    startangle=90,
                    shadow=True,
                    labels=None,
                    colors=color_list,
                    explode=exp_list)
plt.title("Immigration to canada by continet 1980-2013")
plt.axis("equal")
plt.legend(labels=new_df.index,
           loc="upper left")
plt.show()
```

```python
df_t=pd.DataFrame(new_df[years].sum(axis=0))
df_t.head()
```

```python
df_t.index=map(float,df_t.index)
df_t.reset_index(inplace=True)
df_t.head(2)
```

```python
df_t.columns=["year","total"]
df_t.head(2)
```

```python
import seaborn as sns
mpl.style.use("ggplot")
plt.figure(figsize=(15,10))
sns.set(font_scale=1.5)

reg_plot= sns.regplot(x="year",
                     y="total",
                     data=df_t,
                     color="y",
                     marker="+",
                     scatter_kws={"s":200})

reg_plot.set(xlabel="year",
      ylabel="total immigrantion to canada")
reg_plot.set_title("total immigration to canada 1980-2013")
plt.show()
```
