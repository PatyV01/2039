```python
import numpy as np
import pandas as pd

df=pd.read_csv("datasets/antropometria-dataset.csv")
df.head(2)
```

```python
df.columns
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
da=pd.DataFrame
da.hist(data=df,
        column=["sistol","diastol"],
        bins=20)
plt.show()
```

```python
da.hist(data=df,
                 column=["peso","cintura"],
                 bins=50)
plt.show()
```

```python
X=df["peso"].values
y=df["cintura"].values

plt.scatter(X,y,c="g")
plt.show()
        
```

La grafica de cintura la veo simetrica, debido a no existe ninguna cola.
