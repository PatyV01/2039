```python
import numpy as np
import pandas as pd
df=pd.read_excel("datasets/Canada.xlsx",
                sheet_names="Canada by Citizenship",
                skiprows=range(20),
                skipfooter=2,
                engine="openpyxl")
```

```python
df.head(2)
#df.columns
```

```python
df.drop(["AREA","REG","DEV","Type","Coverage"],
        axis=1,
        inplace=True)

```

```python
df.columns
```

```python
df.rename(columns={"OdName":"Country",
                  "AreaName":"Continent",
                  "RegName":"Regions"},
         inplace=True)
df.head()
```

```python
df.set_index("Continent", inplace=True)
```

```python
df.head(2)
```

```python
df["Total"]=df.iloc[:,4:].sum(axis=1)
```

```python
df.head(2)
```

```python
years=list(map(str,range(1980,2014)))
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use("ggplot")
```

```python
df_continets=df.groupby("Continent",
                       axis=0).sum()
```

```python
df_continets
```

```python
#color_list=["gold","yellowgreen","lightcoral","lightskyblue","lightgreen","pink","red"]
#explode_list=[0.1,0,0,0,0.2,0.3]

df_continets["Total"].plot(kind="pie",
                           figsize=(5,6),
                           autopct="%10.1f%%",
                           startangle=90,
                           shadow=True)
plt.show()
```
