```python
import pandas as pd
import numpy as np

df=pd.read_csv('datasets/census.csv')
df.head(2)
```

```python
pd.pivot_table(df,
              index='STNAME',
              values='BIRTHS2010',
              aggfunc= 'sum') #aggragate function = Suma
```

```python
pd.pivot_table(df,
              index='STNAME',
              values='BIRTHS2010',
              columns= 'COUNTY',
               margins= True)
```

```python
df=pd.read_csv('datasets/airline_data.csv',
              low_memory=False)
df.head(3)
```

```python
my_pivot=pd.pivot_table(df,
              index=['Year','Quarter'],
              values='DepDelay',
              columns='DepartureDelayGroups',
              aggfunc= 'sum',
              margins= True)
```

```python
my_pivot.to_csv('mypivot.csv')
```
